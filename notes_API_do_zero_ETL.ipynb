Como engenheiros de dados precisamos ser eficientes. Se queremos a cotação do bitcoin nao faz sentido trazer TODA a informação de uma site que tem a cotação, queremos apenas o valor. 

Automatizar via API sempre é uma opção nas empresas e dá bons projetos com saving, selenium ou outros webscraping podem ser usados mas é menos eficiente, manual só quando absolutamente necessário

Temos 2 tipos de bancos de dados noSQL:
1. Tinydb e mongodb são do tipo objeto - não tem schema rígido e pré-definido
2. redis é do tipo chave-valor
3. vetoriais sao usados por LLMS

SQLite e Tinydb nao suportam multiplos acessos (concurrency)

NoSQL é mais usado quando precisamos de flexibilidade, noSQL normalmente está mais para o lado do 'E' na ETL e SQL está mais no lado 'L' da ETL

- noSQL salva a informação rapido SQL salva devagar (pq tem schema, constains, stored procedures, etc)
- noSQL demora para fazer consulta e SQL é rapido

mySQL (Oracle) é open source mas só para uso próprio, postgres pode vender
