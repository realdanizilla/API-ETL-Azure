Como engenheiros de dados precisamos ser eficientes. Se queremos a cotao do bitcoin nao faz sentido trazer TODA a informao de uma site que tem a cotao, queremos apenas o valor. 

Automatizar via API sempre  uma opo nas empresas e d bons projetos com saving, selenium ou outros webscraping podem ser usados mas  menos eficiente, manual s quando absolutamente necessrio

Temos 2 tipos de bancos de dados noSQL:
1. Tinydb e mongodb so do tipo objeto - no tem schema rgido e pr-definido
2. redis  do tipo chave-valor
3. vetoriais sao usados por LLMS

SQLite e Tinydb nao suportam multiplos acessos (concurrency)

NoSQL  mais usado quando precisamos de flexibilidade, noSQL normalmente est mais para o lado do 'E' na ETL e SQL est mais no lado 'L' da ETL

- noSQL salva a informao rapido SQL salva devagar (pq tem schema, constains, stored procedures, etc)
- noSQL demora para fazer consulta e SQL  rapido

mySQL (Oracle)  open source mas s para uso prprio, postgres pode vender
